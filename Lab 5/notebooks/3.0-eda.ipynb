**Answering this question will** help us to see the current population distribution status and also understand more about current thinking on gender in Asia.

**How we answer this question**: we will compare the data on Female Population and Male Population over the time.

**YOUR ANSWER HERE**

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 
```

### Analyzing to get the answer

```python
# YOUR CODE HERE
population_df = pd.DataFrame()
population_df["Total Population"] = cleaned_df.groupby("Year")["Total Population"].sum()
population_df["Female Population"] = cleaned_df.groupby("Year")["Female Population"].sum()
population_df["Male Population"] = cleaned_df.groupby("Year")["Male Population"].sum()

population_df["Year"] = population_df.index
population_df.reset_index(drop=True, inplace=True)
population_df.sort_values("Total Population", inplace=True, ascending=False)

population_df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
ax = sns.lineplot(data=population_df, x="Year", y= "Total Population", label='Total Population')
ax = sns.lineplot(data=population_df, x="Year", y="Female Population", label="Female Population")
ax = sns.lineplot(data=population_df, x="Year", y="Male Population", label="Male Population")
```

**OBSERVATION AND CONCLUSION**

* In general, Asia's population is growing rapidly, following an upward linear trend, with both men and women experiencing similar growth rates. While the lines representing men and women appear parallel, the male population is larger.

* It seems that Asia's population is still growing for both genders, but at a slower pace. According to natural demographics, the sex ratio should be approximately 1:1. However, the graph displays a gap, indicating that biases towards favoring the birth of male babies over female babies still persist.
